```python
#deklaracja istotnych bibliotek
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

PrzedziaÅ‚ 1. x>-1 oraz x<1


```python
x1 = np.linspace(-1, 1, 100) #PrzedziaÅ‚ 1. x>-1 oraz x<1
x2 = np.linspace(-6, 6, 100) #PrzedziaÅ‚ 2. x>-6 oraz x<6
x3 = np.linspace(0, 5, 100) #PrzedziaÅ‚ 3. x>0 oraz x<5

y = x1**2 + 5 #funkcja

p = [x1, x2, x3]
for i in p:
    plt.plot(i, y, label = 'x1**2 + 5') #zdefinowana legenda
    plt.title("Funkcja liniowa f(x) = x1**2 + 5")
    plt.legend() #wyswietl legande
    plt.xlabel("x") #argumenty osi x
    plt.ylabel("y") #argumenty osi y
    plt.grid() #wyswietl siatke 
    plt.show() #wyswietl
```
